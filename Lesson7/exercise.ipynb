```python
# Задача 1
import pandas as pd
import numpy as np

# Создание DataFrame с 5 столбцами и 10 строками, заполненный случайными числами от 0 до 1
df = pd.DataFrame(np.random.rand(10, 5), columns=['A', 'B', 'C', 'D', 'E'])

# Функция для вычисления среднего по строке для значений больше 0.3
def mean_above_threshold(row, threshold=0.3):
    return row[row > threshold].mean()

# Применение функции к каждой строке
df['MeanAbove0.3'] = df.apply(mean_above_threshold, axis=1)

# Вывод результата
df
```

```python
# Задача 2
# Чтение данных из CSV файла
wells_info = pd.read_csv('wells_info.csv')

# Преобразование дат в формат datetime
wells_info['SpudDate'] = pd.to_datetime(wells_info['SpudDate'])
wells_info['FirstProductionDate'] = pd.to_datetime(wells_info['FirstProductionDate'])

# Вычисление разницы в месяцах между SpudDate и FirstProductionDate
wells_info['MonthsOfProduction'] = ((wells_info['FirstProductionDate'].dt.year - wells_info['SpudDate'].dt.year) * 12 +
                                     (wells_info['FirstProductionDate'].dt.month - wells_info['SpudDate'].dt.month))

# Подсчет целых месяцев
wells_info['WholeMonths'] = wells_info['MonthsOfProduction'].apply(lambda x: max(0, x))

# Вывод результатов
wells_info[['API', 'WholeMonths']]
```

```python
# Задача 3
# Чтение данных из CSV файла с пропущенными значениями
wells_info_na = pd.read_csv('wells_info_na.csv')

# Заполнение пропущенных числовых значений медианой
numeric_cols = wells_info_na.select_dtypes(include=[np.number]).columns
for col in numeric_cols:
    wells_info_na[col].fillna(wells_info_na[col].median(), inplace=True)

# Заполнение остальных пропущенных значений самым частым значением
categorical_cols = wells_info_na.select_dtypes(exclude=[np.number]).columns
for col in categorical_cols:
    wells_info_na[col].fillna(wells_info_na[col].mode()[0], inplace=True)

# Вывод результата
wells_info_na
```
